# Compensating Inter-Dataset Variability in PLDA Hyper-Parameters for Robust Speaker Recognition
补偿PLDA参数的不同数据集差异性, 实现更鲁棒性的spker Recognition

# Abstract 
IDVC -- inter-database variablity compensation  不同数据集间差异补偿方法, 
IDVC 补偿数据集 ivector之间的偏移, 通过约束这些偏移到一个低维子空间中.   
该子空间 通过不同来源数据集, 每个来源被分割到一个子集中 来进行估计.  
在本文 我们推广IDVC方法 来补偿由于PLDA中(within between class)协方差参数带来的不同数据集之间的差异性.  
使用提出的方法 我们设法覆盖了由于PLDA利用不匹配数据训练导致性能降低的 85%. 

本文是在 Domain Adaption Challenge框架下的结果, 本文主要的目标是提升域自适应系统准确率, 通过SWB构建的系统 在 NIST SRE10上进行评估的准确率.  没有任何SRE数据(SRE10之前数据)自适应情况下.   

应对数据集失配的研究挑战, 之前已经有一些方法, 使用一些域内数据自适应来提高性能.   
对比这些方法, IDVC 和 声道归整SN 技术不需要自适应数据. 
IDVC 目标是显式的在 ivector子空间中建模 数据集偏移差异, 然后在一个步骤中补偿这些数据差异, IDVC 性能比 SN更好.
在本文中 我们研究了各种不匹配的类型而不是简单的ivector空间的偏移, 然后推广IDVC 能够处理数据集变化.  我们汇报一个经验结果, 表明补偿这些数据集失配下参数变化, 来提高性能, 实现更高鲁棒性的性能.

# 3 Inter Database Variability Compensation
不匹配数据集间 差异补偿 
IDVC 目的是估计 不匹配数据集在ivector空间的偏移, 然后在ivector子空间正确的去除不匹配数据间的偏移.
实现方法:   
1 将开发集 按照信道分为多个不匹配子集, 每个训练一个 PLDA模型   
2 在不同信道子集下的PLDA模型参数(Sb Sw) 之间的差异, 实际上就代表了不匹配数据集的相对偏移, 因此需要分析不匹配数据PLDA模型的参数差异.  
3 然后对于这些参数差异, 估计出一个 ivector空间的 低维子空间.    
4 然后估计出来的 低维子空间 被从 ivector空间中移除, (注意要在 length-normalize  LDA 之前).  
下面详细描述:  

## 3.2 Motivation for proposed method
我们假设 ivector空间中 有些方向 对于 信道失配更佳敏感,  为了实现一个 对信道失配数据更鲁棒的PLDA系统,  我们需要找到并且移除那些对信道失配更敏感方向构成的 低维子空间.   
这个低维子空间, 叫做 失配敏感子空间.    


失配敏感子空间 可以从 多个纯净信道计算得到的PLDA模型集合中估计出来.   

我们希望(结果也是如期望) 从开发集估计出来的失配敏感子空间 可以推广到 未见的评估数据上.  

## 3.3 outline of the proposed method
提出方法大纲

+ A IDVC 训练 不同信道数据子集差异空间估计:  
1 分割开发数据集为 多个 纯净信道子集  
2 为每个子集i 估计PLDA 模型, 得到每个子集 PLDA 参数$\{\mu_i, B_i, W_i\}$   
3 为所有子集的 $\{\mu_i\}$ 估计一个 ivector空间里面对$\mu_i$ 敏感的子空间 $\{S_{\mu}\}$    
4 为所有子集的 $\{B_i\}$ 估计一个 ivector空间里面对$B_i$ 敏感的子空间 $\{S_{B}\}$   
5 为所有子集的 $\{W_i\}$ 估计一个 ivector空间里面对$W_i$ 敏感的子空间 $\{S_{W}\}$    
6 联合这些子空间 形成一个单一子空间 $S = S_{\mu} + S_{W} + S_{B} $   
完成 IDVC训练   

+ B PLDA 训练
1 利用IDVC结果$ S $, 从 ivectors 中移除 信道影响子空间 $S$, 
2 使用处理过的ivectors 训练PLDA

+ C PLDA 打分
1 从 评估集合ivectors 移除 $S$   
2 使用PLDA打分

# 5 Discussion

标准的PLDA处理了 两个类型的方差信息: Between variance  Within variance,  Gaussion PLDA 当前是最佳的模型假设, 两个方差满足多元高斯分布.   
对于混杂数据, 高斯假设又是不合适的, 然而真实数据常常是混在数据,(不同信道, 不同音频时长, 不同内容信息).  
Source Normalized (SN 信道归整) and IDVC方法 都是处理不同信道数据偏移的方法,  

SN 估计了不通信道 ivectors的标记数据 间的协方差, 然后将 不同信道间的差异 从 Between class 协方差 传递到了 Within class 协方差中, 因此实现将信道失配的影响, 不会影响分类信息.   
IDVC 方法实际上估计了 相同的 ivectors空间内的 信道协方差, 然后从ivector空间直接移除了, 因此 IDVC比 SN更佳有效, 当评估集与开发集更佳不匹配的时候.   


**在本文中, 我们将原始的IDVC方法扩展到了 信道失配数据集(能够导致ivector空间的一个偏移).  
我们希望在ivector空间中 找到一些方向, 这些方向在 Between Within class方差上 会对 不同信道 很敏感, 这些方向 对不同的信道上 在 Sb Sw 会有很大方差, 这些实际上是信道信息, 因此要从 ivector空间中去除, 这样 才能更好的保证 Sb 只受到 class 信息影响.**
